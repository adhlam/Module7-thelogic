# **The Logic**

Let's use the following three propositions:


p: "It is Monday."

q: "I have a meeting."

r: "I woke up early."

I will assume that p and r are true, but q can be either true or false, depending on whether I have actually have a meeting that day or not.

```python
ItisMonday = True
IHaveAMeeting = False
IWokeUpEarly = True
```

# $$\lnot q$$

```python
not(IHaveAMeeting)
```

# $$r \land p$$

**r  p: "I woke up early and it is Monday."**

```python
IWokeUpEarly and ItisMonday
```

# $r\Rightarrow p$

**r  p: "If I woke up early, then it is Monday."**

```python
def implies(r,p):
  return r or p

implies(IWokeUpEarly,ItisMonday)
```

# $$ r \lor \lnot q$$

**r  q: "I woke up early or I don't have a meeting."**

```python
def implies(r,q):
  return r or not(q)

implies(IWokeUpEarly,IHaveAMeeting)
```

**Computing the truth values**

```python
import itertools

propositions = ['p', 'q', 'r']

def conjunction(r, p):
    return r and p

def disjunction(r, q):
    return r or not q

def implication(r, p):
    return not r or p

truth_values = list(itertools.product([True, False], repeat=3))

for p, q, r in truth_values:
    result_conjunction = conjunction(r, p)
    result_disjunction = disjunction(r, q)
    result_implication = implication(r, p)
    print(f"p: {p}, q: {q}, r: {r}, (r  p): {result_conjunction}, (r  q): {result_disjunction}, (r  p): {result_implication}")
```

# **Constructing the truth table**

# ** (r  p)  (p  q)**

```python
import itertools

p = [True, False]
q = [True, False]
r = [True, False]

print("p  q  r  rp (rp) pq (rp)(pq)")

for combination in itertools.product(p, q, r):
  
    p_val, q_val, r_val = combination
    
    rp = r_val ^ p_val
    not_rp = not rp
    porq = p_val or q_val
    imp = not_rp <= porq
    
    print("{}  {}  {}  {}    {}     {}    {}".format(int(p_val), int(q_val), int(r_val), int(rp), int(not_rp), int(porq), int(imp)))
```
